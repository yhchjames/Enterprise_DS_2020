```python
page = requests.get('https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Fallzahlen.html')
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
soup.get_text()
```

```python
html_table = soup.find('table') #get the things between table, have to check the sourse code of the website
                                # the differencd between find and findall is that find only return one result
```

```python
all_rows = html_table.find_all('tr') #get all rows in the table, and separate them by <tr>
print(all_rows)
```

```python
final_data_list = []
```

```python
for pos,rows in enumerate(all_rows): #enunetate=>  [(0,dsafdsa),(1,fdsa),(2,fdsaee) ....] 
#     print(pos)
#     print(rows)
    col_list = [each_col.get_text() for each_col in rows.find_all('td')]
        #name and number are in the midle of td
        #get_text onle take the things between the tags
    final_data_list.append(col_list)
print(final_data_list)
```

```python
pd.DataFrame(final_data_list).dropna().rename(columns={0:'state',
                                                      1:'cases',
                                                      2:'changes',
                                                      3:'cases_per_100k',
                                                      4:'fatal'}) #dropna = drop all-zero rows 
```
